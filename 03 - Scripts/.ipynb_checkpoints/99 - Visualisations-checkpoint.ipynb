### 3.1.1. Bar - Day of the Week

```python
# Define the mapping of numeric days to weekday names in chronological order
day_mapping = {0: "Saturday", 1: "Sunday", 2: "Monday", 3: "Tuesday", 
               4: "Wednesday", 5: "Thursday", 6: "Friday"}

# Count occurrences of each day of the week
order_counts = df_dataset['order_day_of_week'].value_counts().reset_index()
order_counts.columns = ['order_day_of_week', 'count']

# Sort by the numeric order of days (0-6) before mapping names
order_counts = order_counts.sort_values(by='order_day_of_week')
order_counts['order_day_of_week'] = order_counts['order_day_of_week'].map(day_mapping)

# Define the order for the x-axis to ensure correct weekday sequence
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Create figure and axis
plt.figure(figsize=(10, 6))

# Plot the bar chart
sns.barplot(data=order_counts, x='order_day_of_week', y='count', order=weekday_order)

# Set the title and labels
plt.title('Distribution of Orders by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Products Ordered')

# Add gridlines corresponding to y-axis values
plt.grid(axis='y', linestyle='dotted', alpha=0.7)

# Format y-axis with thousands separator
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))

# Adjust layout for better spacing
plt.tight_layout()

# Save the figure before displaying
save_path = os.path.join(path, '04 - Analysis/Visualisations', '3.1.1_order_dow.png')
plt.savefig(save_path, dpi=300, bbox_inches='tight')

# Show the plot
plt.show()

# Print confirmation message
print(f"Chart successfully saved at: {save_path}")
```

### 3.1.3. Stacked Area

```python
# Define the mapping of numeric days to weekday names in chronological order
day_mapping = {0: "Saturday", 1: "Sunday", 2: "Monday", 3: "Tuesday", 
               4: "Wednesday", 5: "Thursday", 6: "Friday"}

# Define the correct weekday order (Monday at the bottom, Sunday at the top)
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Count occurrences for each combination of day of week and hour of day
order_counts = df_dataset.groupby(['order_day_of_week', 'order_hour_of_day']).size().reset_index(name='count')

# Map numeric day to weekday names
order_counts['order_day_of_week'] = order_counts['order_day_of_week'].map(day_mapping)

# Pivot the data to get days as columns and hours as rows
order_pivot = order_counts.pivot(index='order_hour_of_day', columns='order_day_of_week', values='count')

# Reorder the columns to match the correct weekday order
order_pivot = order_pivot[weekday_order]

# Define fig and ax properly
fig, ax = plt.subplots(figsize=(12, 6))  # Creates the figure and axis

# Generate a blue gradient color palette with 7 shades (one for each day)
colors = sns.color_palette("Blues", n_colors=len(weekday_order))

# Plot the area chart with the custom color palette
order_pivot.plot(kind='area', stacked=True, color=colors, alpha=0.8, ax=ax)

# Add gridlines corresponding to y-axis values
ax.grid(axis='y', linestyle='dotted', alpha=0.7)

# Format y-axis with thousands separator
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))

# Labels and title
ax.set_title("Distribution of Orders by Time of Day and Day of Week")
ax.set_xlabel("Hour of the Day")
ax.set_ylabel("Number of Products Ordered")
ax.legend(title="Day of Week", loc="upper left")

# Adjust layout for better spacing
plt.tight_layout()

# Save the figure before displaying
save_path = os.path.join(path, '04 - Analysis/Visualisations', '3.1.3_order_area.png')
plt.savefig(save_path, dpi=300, bbox_inches='tight')

# Show the chart
plt.show()

# Print confirmation message
print(f"Chart successfully saved at: {save_path}")
```

### 3.3.3. Box and Whisker

```python
# Calculate key statistics
q1 = np.percentile(df_dataset['prices'], 25)  # 25th percentile
q3 = np.percentile(df_dataset['prices'], 75)  # 75th percentile
median = np.median(df_dataset['prices'])  # Median
mean = np.mean(df_dataset['prices'])  # Mean
iqr = q3 - q1  # Interquartile range

# Define outlier thresholds
lower_whisker = q1 - 1.5 * iqr
upper_whisker = q3 + 1.5 * iqr

# Identify outliers
outliers = df_dataset[(df_dataset['prices'] < lower_whisker) | (df_dataset['prices'] > upper_whisker)]['prices']

# Create figure
plt.figure(figsize=(10, 6))

# Plot boxplot
sns.boxplot(x=df_dataset['prices'], width=0.5, fliersize=5, linewidth=1.5)

# Add annotations for IQR, mean, and outliers
plt.axvline(q1, color='red', linestyle='dashed', label=f'Q1 (25th %ile): {q1:.2f}')
plt.axvline(q3, color='red', linestyle='dashed', label=f'Q3 (75th %ile): {q3:.2f}')
plt.axvline(mean, color='blue', linestyle='dotted', label=f'Mean: {mean:.2f}')

# Label outliers (only if a few, for readability)
if len(outliers) <= 10:  
    for outlier in outliers:
        plt.text(outlier, 0.1, f"{outlier:.2f}", ha='center', va='bottom', fontsize=10, color="black", rotation=90)

# Set title and labels
plt.title('Box Plot of Product Prices')
plt.xlabel('Product Price')

# Format x-axis with thousands separator
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))

# Add gridlines
plt.grid(axis='x', linestyle='dotted', alpha=0.7)

# Show legend
plt.legend()

# Improve layout
plt.tight_layout()

# Save the figure before displaying
save_path = os.path.join(path, '04 - Analysis/Visualisations', '3.3.3_price_sales_boxplot.png')
plt.savefig(save_path, dpi=300, bbox_inches='tight')

# Show the plot
plt.show()

# Print confirmation message
print(f"Chart successfully saved at: {save_path}")
```

```python
# Define the mapping of numeric days to weekday names in chronological order
day_mapping = {0: "Saturday", 1: "Sunday", 2: "Monday", 3: "Tuesday", 
               4: "Wednesday", 5: "Thursday", 6: "Friday"}

# Define the order for the x-axis to ensure correct weekday sequence
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Define order for activity_flag categories
activity_order = ["Low Activity", "Moderate Activity", "High Activity"]

# Count occurrences of each day of the week grouped by loyalty_flag
order_counts = df_dataset.groupby(['order_day_of_week', 'loyalty_flag']).size().reset_index(name='count')

# Sort by numeric order of days before mapping names
order_counts['order_day_of_week'] = order_counts['order_day_of_week'].map(day_mapping)

# Pivot the data to structure it for stacking
order_pivot = order_counts.pivot(index='order_day_of_week', columns='loyalty_flag', values='count').fillna(0)

# Reindex to ensure correct weekday order
order_pivot = order_pivot.reindex(weekday_order)

# Define shades of blue for stacking
colors = ["#9ecae1", "#6baed6", "#2171b5"]  # Light to dark blue

# Create figure and axis
plt.figure(figsize=(10, 6))

# Plot stacked bar chart
order_pivot.plot(kind='bar', stacked=True, color=colors, figsize=(10, 6))

# Set labels and title
plt.title('Distribution of Orders by Day of the Week (Stacked by Loyalty Flag)')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Products Ordered')

# Format y-axis with thousands separator
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))

# Add legend with proper positioning
plt.legend(title="Loyalty Flag", loc='upper left', bbox_to_anchor=(1, 1))

# Adjust layout for better spacing
plt.tight_layout()

# Save the figure before displaying
#save_path = os.path.join(path, '04 - Analysis/Visualisations', '3.1.1_order_dow_stacked.png')
#plt.savefig(save_path, dpi=300, bbox_inches='tight')

# Show the plot
plt.show()

# Print confirmation message
#print(f"Chart successfully saved at: {save_path}")
```

```python
# Calculate the percentage of products in each department for each loyalty category
dept_loyalty_counts = df_dataset.groupby(['department', 'price_label']).size().unstack(fill_value=0)

# Convert counts to percentages within each department
dept_loyalty_percentages = dept_loyalty_counts.div(dept_loyalty_counts.sum(axis=1), axis=0) * 100

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Define custom blue shades
blue_shades = ["#08306b", "#2171b5", "#6baed6", "#c6dbef"]

# Plot a stacked bar chart
dept_loyalty_percentages.plot(kind='bar', stacked=True, ax=ax, color=blue_shades)

# Set title and labels
plt.title('Percentage Distribution of Products by Department and Loyalty Category')
plt.xlabel('Department')
plt.ylabel('Percentage of Products')

# Format y-axis as percentage
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))

# Add dotted gridlines on the y-axis
plt.grid(axis='y', linestyle='dotted', alpha=0.7)

# Improve layout for readability
plt.xticks(rotation=45, ha='right')
plt.legend(title="Loyalty Flag", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Show the plot
plt.show()
```
